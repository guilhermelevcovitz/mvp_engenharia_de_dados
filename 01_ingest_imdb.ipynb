```python
from pyspark.sql import functions as F

basics_path = "/Volumes/workspace/default/imdb_raw/title.basics.tsv.gz"
ratings_path = "/Volumes/workspace/default/imdb_raw/title.ratings.tsv.gz"

df_basics = (
    spark.read
         .option("header", True)
         .option("sep", "\t")
         .option("nullValue", "\\N")     
         .csv(basics_path)
)

df_ratings = (
    spark.read
         .option("header", True)
         .option("sep", "\t")
         .option("nullValue", "\\N")
         .csv(ratings_path)
)

print("BASICS:")
display(df_basics.limit(5))

print("RATINGS:")
display(df_ratings.limit(5))

bronze_base = "/Volumes/workspace/default/imdb_bronze"

dbutils.fs.rm(bronze_base, recurse=True)

df_basics.write.format("delta").mode("overwrite").save(bronze_base + "/title_basics")
df_ratings.write.format("delta").mode("overwrite").save(bronze_base + "/title_ratings")

print("Bronze salvo em:", bronze_base)
```
